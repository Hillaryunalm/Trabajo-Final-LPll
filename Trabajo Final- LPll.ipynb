```python
def obtener_datos(producto):
    info_producto = []
    for producto in productos:
        # nombre
        nombre = producto.find_elements("xpath",'.//a[@class="Showcase__name"]')
        nombre= [i.text for i in nombre]
        # contenido
        contenido = producto.find_elements("xpath",'.//div[@class="Showcase__units-reference"]')
        contenido=[i.text for i in contenido]
        # precio
        precio = producto.find_elements("xpath",'.//span[@class="price"]')
        precio=[i.text for i in precio]
        # Armamos un diccionario para almacenar los datos
        data_dict={
            'nombre':nombre,
            'contenido':contenido,
            'precio':precio}
        info_producto.append(data_dict)
        
    return info_producto
```
